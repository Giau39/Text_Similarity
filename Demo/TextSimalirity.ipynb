```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize
from nltk.stem import *
from stop_words import get_stop_words
from underthesea import word_tokenize, ner
import re

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt', download_dir='/usr/share/nltk_data/tokenizers')
```

```python
# Text preprocessing
stop_words_en = set(stopwords.words('english'))
porter_stemmer_en = PorterStemmer()
wordnet_lemmatizer_en = WordNetLemmatizer()

stop_words_vi = get_stop_words("vi")

def preprocess_text(text, language='en'):
    # Convert to lowercase
    text = text.lower()

    # Tokenization
    if language == 'en':
        text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
        words = text.split()
        words = [porter_stemmer_en.stem(word) for word in words if word not in stop_words_en]
        words = [wordnet_lemmatizer_en.lemmatize(word) for word in words]
    elif language == 'vi':
        text = re.sub(r'[^\w\s]', '', text)
        tokens = word_tokenize(text)
        tokens = [word for word in tokens if word not in stop_words_vi]
        words = [ner(word)[0][0] for word in tokens]

    # Join the processed words back into a sentence
    text = ' '.join(words)

    return text

# Áp dụng tiền xử lý cho cột 'en'
df['en'] = df['en'].apply(lambda x: preprocess_text(x, language='en'))

# Áp dụng tiền xử lý cho cột 'vi'
df['vi'] = df['vi'].apply(lambda x: preprocess_text(x, language='vi'))

# Hiển thị kết quả
display(df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from transformers import BertTokenizer, BertForSequenceClassification, AdamW, BertModel
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
import torch
```

```python
# Chọn một mô hình đã được đào tạo sẵn, ở đây sử dụng mBERT
model_name = "bert-base-multilingual-cased"
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=2, output_hidden_states=True)
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python
# Tải dữ liệu từ các tệp CSV
train_data = pd.read_csv('train_data.csv')
test_data = pd.read_csv('test_data.csv')
dev_data = pd.read_csv('dev_data.csv')

# Chia tập dữ liệu thành features và labels
train_texts, train_labels = train_data[['en', 'vi']].astype(str).values.tolist(), train_data['label'].tolist()
dev_texts, dev_labels = dev_data[['en', 'vi']].astype(str).values.tolist(), dev_data['label'].tolist()
```

```python
# Token hóa dữ liệu
train_encodings = tokenizer(train_texts, truncation=True, padding=True)
dev_encodings = tokenizer(dev_texts, truncation=True, padding=True)

# Tạo Dataset cho PyTorch
class CustomDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

train_dataset = CustomDataset(train_encodings, train_labels)
dev_dataset = CustomDataset(dev_encodings, dev_labels)

# Tạo DataLoader cho PyTorch
batch_size = 16
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
dev_loader = DataLoader(dev_dataset, batch_size=batch_size, shuffle=False)

# Sử dụng GPU nếu có
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Tiến hành fine-tuning
optimizer = AdamW(model.parameters(), lr=2e-5)
num_epochs = 3

# Cập nhật learning rate theo số bước
total_steps = len(train_loader) * num_epochs
scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr=2e-5, steps_per_epoch=len(train_loader), epochs=num_epochs)

for epoch in range(num_epochs):
    model.train()
    for batch in tqdm(train_loader, desc=f"Epoch {epoch + 1}"):
        inputs = {key: val.to(device) for key, val in batch.items()}
        optimizer.zero_grad()
        outputs = model(**inputs)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        scheduler.step()  # Cập nhật learning rate
```

```python
# Đánh giá trên tập validation
model.eval()
with torch.no_grad():
    dev_loss = 0.0
    predictions = []
    true_labels = []
    for batch in tqdm(dev_loader, desc="Validation"):
        inputs = {key: val.to(device) for key, val in batch.items()}
        outputs = model(**inputs)
        loss = outputs.loss
        logits = outputs.logits
        dev_loss += loss.item()
        preds = torch.argmax(logits, dim=1)
        predictions.extend(preds.cpu().numpy())
        true_labels.extend(inputs['labels'].cpu().numpy())

    accuracy = accuracy_score(true_labels, predictions)
    precision = precision_score(true_labels, predictions)
    recall = recall_score(true_labels, predictions)
    f1 = f1_score(true_labels, predictions)
    average_dev_loss = dev_loss / len(dev_loader)

# Tạo DataFrame với thông số đánh giá
evaluation_results = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score", "Validation Loss"],
    "Value": [accuracy, precision, recall, f1, average_dev_loss]
})

# Hiển thị DataFrame dưới dạng bảng
display(evaluation_results)
```
